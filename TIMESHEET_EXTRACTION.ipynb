```python
import os
import pdfplumber
import pandas as pd
from google.colab import files

#  Step 1: Identify Uploaded PDF File
current_path = "/content/"
uploaded_files = os.listdir(current_path)

# Find the most recent PDF file
pdf_files = [f for f in uploaded_files if f.lower().endswith(".pdf")]
if not pdf_files:
    raise FileNotFoundError(" No PDF file found in /content/. Please upload your timesheet PDF.")

pdf_path = os.path.join(current_path, pdf_files[-1])  # Use the latest uploaded PDF
print(f" Using file: {pdf_path}")

#  Step 2: Extract Data from the Timesheet PDF
employee_data = []

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        # Extract text from the page
        text = page.extract_text()

        if text:
            lines = text.split("\n")  # Split text into individual lines

            # Initialize variables for employee details
            employee_name, location, cost_center, allocation_percentage, hours_reported = None, None, None, None, None

            for line in lines:
                line = line.strip()

                # Extract Employee Name
                if "Employee Name:" in line:
                    employee_name = line.split("Employee Name:")[-1].strip()

                # Extract Location
                if "Branch Office Name :" in line:
                    location = line.split("Branch Office Name :")[-1].strip()

                # Extract Cost Center, Allocation %, and Hours Reported
                if "%" in line and "Total" not in line:  # Avoid summary rows
                    parts = line.split()
                    if len(parts) >= 3:
                        cost_center = parts[-3]  # Assuming cost center appears near the end
                        allocation_percentage = parts[-2].replace("%", "")  # Remove '%' sign
                        hours_reported = parts[-1]  # Extract hours worked

                        # Append data to list
                        employee_data.append({
                            "Employee Name": employee_name,
                            "Location": location,
                            "Cost Center": cost_center,
                            "Allocation %": allocation_percentage,
                            "Hours Reported": hours_reported
                        })

#  Step 3: Convert Extracted Data to DataFrame
df_summary = pd.DataFrame(employee_data)

#  Step 4: Save the Extracted Data to an Excel File
output_file = os.path.join(current_path, "Timesheet_Report_Period_8.xlsx")
df_summary.to_excel(output_file, index=False)

#  Step 5: Provide File for Download
files.download(output_file)  # Automatically download the file
print(f" Report saved as: {output_file}")

#  Step 6: Display Extracted Data in Google Colab
import ace_tools as tools
tools.display_dataframe_to_user(name="Employee Cost Center Summary", dataframe=df_summary)
```

```python
import pdfplumber
import pandas as pd

# Define file path
pdf_path = "/content/2025-01-31 MCC VT Staff Time sheets Period 8 - updated.pdf"

# Initialize a list to store extracted data
employee_data = []

# Define the branch office name
branch_office = "ECDC MCC Brattleboro, VT"

# Extract data from PDF
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            lines = text.split("\n")

            # Initialize variables
            employee_name = None
            ssn_last4 = None
            location = None

            for i, line in enumerate(lines):
                line = line.strip()

                # Extract Employee Name and SSN (they are merged in one line)
                if "Employee Name:" in line and "Social Security Number:" in line:
                    parts = line.split("Social Security Number:")
                    employee_name = parts[0].replace("Employee Name:", "").strip()
                    ssn_last4 = parts[1].strip()

                # Extract Location (75 or 77)
                if "LOC" in line and "%" in lines[i + 1]:  # Finding the LOC line
                    parts = lines[i + 1].split()
                    if len(parts) >= 2:
                        location = parts[0]  # First LOC value

                # Extract Funder, Grant Name, Cost Center (Purpose Code), Allocation %, and HRS Split
                if "%" in line and "Total" not in line:
                    parts = line.split()
                    if len(parts) >= 6:
                        funder = parts[1]  # Funder number
                        grant_name = " ".join(parts[2:-3])  # Grant name
                        cost_center_purpose_code = parts[-3]  # Merging Cost Center and Purpose Code
                        allocation = parts[-2].replace("%", "")  # Allocation percentage
                        hrs_split = parts[-1]  # Hours split

                        # Append data only if it's a valid entry (avoid headers repeating per user)
                        if funder.isnumeric() and allocation.replace(".", "").isdigit():
                            employee_data.append({
                                "Employee Name": employee_name,
                                "SSN (Last 4 Digits)": ssn_last4,
                                "Branch Office": branch_office,
                                "Location (LOC)": location,
                                "Funder": funder,
                                "Grant Name": grant_name,
                                "Cost Center (Purpose Code)": cost_center_purpose_code,  # Merged column
                                "% Allocated": allocation,
                                "HRS Split": hrs_split
                            })

# Convert to DataFrame
df_employee_summary = pd.DataFrame(employee_data)

# Summarize total hours per employee
df_employee_summary["HRS Split"] = pd.to_numeric(df_employee_summary["HRS Split"], errors='coerce')
df_total_hours = df_employee_summary.groupby(["Employee Name", "SSN (Last 4 Digits)"], as_index=False)["HRS Split"].sum()
df_total_hours.rename(columns={"HRS Split": "Total Hours"}, inplace=True)

# Merge the total hours back into the main dataset
df_employee_summary = df_employee_summary.merge(df_total_hours, on=["Employee Name", "SSN (Last 4 Digits)"], how="left")

# Remove duplicate rows caused by repeating headers per employee
df_employee_summary = df_employee_summary.drop_duplicates()

# Save the extracted data to an Excel file
output_file = "/content/Final_Summary_Period_8.xlsx"
df_employee_summary.to_excel(output_file, index=False)

# Display the extracted data
from IPython.display import display
display(df_employee_summary)  # This will show the table in Colab

#  Provide file for download
from google.colab import files
output_file = "/content/Final_Summary_Period_8.xlsx"
df_employee_summary.to_excel(output_file, index=False)

#  Download the file
files.download(output_file)
```

```python
import pdfplumber
import pandas as pd
import re

# Define file path for Google Colab
pdf_path = "/content/2025-01-31 MCC VT Staff Time sheets Period 8 - updated.pdf"

# Initialize a list to store extracted data
employee_data = []

# Define the branch office name
branch_office = "ECDC MCC Brattleboro, VT"

# Extract data from PDF
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            lines = text.split("\n")

            # Initialize variables
            employee_name = None
            ssn_last4 = None
            location = None
            total_hours = None

            for i, line in enumerate(lines):
                line = line.strip()

                # Extract Employee Name and SSN (they are merged in one line)
                if "Employee Name:" in line and "Social Security Number:" in line:
                    parts = line.split("Social Security Number:")
                    employee_name = parts[0].replace("Employee Name:", "").strip()
                    ssn_last4 = parts[1].strip()

                # Extract Location (75 or 77)
                if "LOC" in line and "%" in lines[i + 1]:
                    parts = lines[i + 1].split()
                    if len(parts) >= 2:
                        location = parts[0]

                # Extract Total Hours from W1 + W2
                if "W1 + W2" in line or "TOTAL" in line:
                    total_hours_match = re.findall(r"\d+\.\d+|\d+", lines[i + 1])
                    total_hours = float(total_hours_match[-1]) if total_hours_match else None

                # Extract Funder, Grant Name, Cost Center (Purpose Code), Allocation %, and HRS Split
                if "%" in line and "Total" not in line:
                    parts = line.split()
                    if len(parts) >= 7:
                        funder = parts[1]  # Extract the first 3 digits as Funder Code
                        grant_name = " ".join(parts[2:-4])
                        cost_center_purpose_code = parts[-4]
                        allocation = re.sub("[^0-9.]", "", parts[-3])

                        # Compute HRS Split safely
                        try:
                            allocation_value = float(allocation) if allocation else 0
                            hrs_split = (total_hours * (allocation_value / 100)) if total_hours else 0
                        except ValueError:
                            allocation_value = 0
                            hrs_split = 0

                        # Append extracted data
                        employee_data.append({
                            "Employee Name": employee_name,
                            "SSN (Last 4 Digits)": ssn_last4,
                            "Branch Office": branch_office,
                            "Location (LOC)": location,
                            "Funder": funder,
                            "Grant Name": grant_name,
                            "Cost Center (Purpose Code)": cost_center_purpose_code,
                            "% Allocated": allocation_value,
                            "HRS Split": round(hrs_split, 2),
                            "Total Hours": total_hours
                        })

# Convert to DataFrame
df_employee_summary = pd.DataFrame(employee_data)

# Convert % Allocated, HRS Split, and Total Hours to numeric format
df_employee_summary["% Allocated"] = pd.to_numeric(df_employee_summary["% Allocated"], errors='coerce')
df_employee_summary["HRS Split"] = pd.to_numeric(df_employee_summary["HRS Split"], errors='coerce')
df_employee_summary["Total Hours"] = pd.to_numeric(df_employee_summary["Total Hours"], errors='coerce')

# Save the corrected report
output_file = "/content/Final_Corrected_Timesheet_Report_Period_8.xlsx"
df_employee_summary.to_excel(output_file, index=False)

#  Provide file for download
from google.colab import files
output_file = "/content/Final_Summary_Period_8.xlsx"
df_employee_summary.to_excel(output_file, index=False)

#  Download the file
files.download(output_file)
```

```python
import pdfplumber
import pandas as pd

# Define file path
pdf_path = "/content/2025-01-31 MCC VT Staff Time sheets Period 8 - updated.pdf"

# Initialize a list to store extracted data
employee_data = []

# Define the branch office name
branch_office = "ECDC MCC Brattleboro, VT"

# Extract data from PDF
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            lines = text.split("\n")

            # Initialize variables
            employee_name = None
            ssn_last4 = None
            total_hours = None

            for i, line in enumerate(lines):
                line = line.strip()

                # Extract Employee Name and SSN (they are merged in one line)
                if "Employee Name:" in line and "Social Security Number:" in line:
                    parts = line.split("Social Security Number:")
                    employee_name = parts[0].replace("Employee Name:", "").strip()
                    ssn_last4 = parts[1].strip()

                # Extract Total Hours
                if "W1 + W2" in line and "TOTAL" in lines[i + 1]:
                    total_hours = lines[i + 2].split()[-1]  # Get the total hours from the last column

                # Extract allocation details
                if "Location #" in line and "Funder #" in line:
                    for j in range(i + 2, len(lines)):
                        detail_line = lines[j].strip()

                        # Stop if we reach the total row or another employee
                        if "Total" in detail_line or "Employee Name:" in detail_line:
                            break

                        parts = detail_line.split()
                        if len(parts) >= 6:
                            location = parts[0]
                            funder = parts[1]
                            grant_name = " ".join(parts[2:-3])
                            cost_center_purpose_code = parts[-3]
                            allocation = parts[-2].replace("%", "").strip()
                            hrs_split = parts[-1]

                            # Convert allocation and calculate HRS Split
                            allocation_value = float(allocation) if allocation.replace(".", "").isdigit() else 0.0
                            total_hours_value = float(total_hours) if total_hours and total_hours.replace(".", "").isdigit() else 0.0
                            hrs_split_value = (total_hours_value * allocation_value / 100) if total_hours_value else 0.0

                            # Append data
                            employee_data.append({
                                "Employee Name": employee_name,
                                "SSN (Last 4 Digits)": ssn_last4,
                                "Branch Office": branch_office,
                                "Location (LOC)": location,
                                "Funder": funder,
                                "Grant Name": grant_name,
                                "Cost Center (Purpose Code)": cost_center_purpose_code,
                                "% Allocated": allocation_value,
                                "HRS Split": hrs_split_value,
                                "Total Hours": total_hours_value
                            })

# Convert to DataFrame
df_employee_summary = pd.DataFrame(employee_data)

#  Provide file for download
from google.colab import files
output_file = "/content/Final_Summary_Period_8.xlsx"
df_employee_summary.to_excel(output_file, index=False)

#  Download the file
files.download(output_file)
```
